## è‡ªè¡Œå®šç¾©å‡½æ•¸ `calculate_circle` èƒ½å¤ å°‡è¼¸å…¥çš„åŠå¾‘ä¾ç…§åƒæ•¸ `to` è¨ˆç®—ç‚ºé¢ç©æˆ–å‘¨é•·ï¼Œä½¿ç”¨ $\pi = 3.14$ã€‚

\begin{equation}
Area = \pi r^2 \\
Perimeter = 2 \pi r \\
\text{where} \; \pi = 3.14
\end{equation}

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€ä¸€å€‹æ–‡å­—å‘é‡ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡ã€‚

```r
calculate_circle <- function(r, to) {
    # calculate_circle(3, to='area')
    # [1] 28.26
    # calculate_circle(3, to='perimeter')
    # [1] 18.84
    # calculate_circle(4, to='area')
    # [1] 50.24
    # calculate_circle(4, to='perimeter')
    # [1] 25.12
    ### BEGIN SOLUTION
    my_pi <- 3.14
    if (to == 'area') {
        return(my_pi*r**2)
    } else {
        return(2*my_pi*r)
    }
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `convert_degree` èƒ½å¤ å°‡è¼¸å…¥çš„æº«åº¦åº¦æ•¸ä¾ç…§åƒæ•¸ `from` èˆ‡ `to` åœ¨æ”æ°èˆ‡è¯æ°ä¹‹é–“è½‰æ›ã€‚

\begin{equation}
Fahrenheit_{(^{\circ} F)} = Celsius_{(^{\circ} C)} \times \frac{9}{5} + 32 \\
Celsius_{(^{\circ} C)} =  (Fahrenheit_{(^{\circ} F)} - 32) \times \frac{5}{9}
\end{equation}

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€å…©å€‹æ–‡å­—å‘é‡ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€‚

```r
convert_degree <- function(x, from, to) {
    # convert_degree(100, from='Celsius', to='Fahrenheit')
    # [1] 212.0
    # convert_degree(0, from='Celsius', to='Fahrenheit')
    # [1] 32.0
    # convert_degree(212, from='Fahrenheit', to='Celsius')
    # [1] 100.0
    # convert_degree(32, from='Fahrenheit', to='Celsius')
    # [1] 0.0
    ### BEGIN SOLUTION
    if (from == 'Celsius') {
        return(x*9/5 + 32)
    } else {
        return((x - 32)*5/9)
    }
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `get_bmi_category` èƒ½å¤ å°‡è¼¸å…¥çš„èº«é«˜ï¼ˆå…¬åˆ†ï¼‰ã€é«”é‡ï¼ˆå…¬æ–¤ï¼‰æ›ç®—ç‚º BMI é¡åˆ¥ã€‚

\begin{equation}
BMI = \frac{weight_{kg}}{height_{m}^2}
\end{equation}

<https://en.wikipedia.org/wiki/Body_mass_index>

- é æœŸè¼¸å…¥ï¼šå…©å€‹æ•¸å€¼å‘é‡ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡ã€‚

```r
get_bmi_category <- function(height, weight) {
    # get_bmi_category(170, 50)
    # [1] 'Underweight'
    # get_bmi_category(170, 60)
    # [1] 'Normal'
    # get_bmi_category(206, 113)
    # [1] 'Overweight'
    # get_bmi_category(216, 147)
    # [1] 'Obese'
    ### BEGIN SOLUTION
    bmi <- weight / (height*0.01)**2
    if (bmi > 30) {
        category <- 'Obese'
    } else if (bmi >= 25) {
        category <- 'Overweight'
    } else if (bmi >= 18.5) {
        category <- 'Normal'
    } else {
        category <- 'Underweight'
    }
    return(category)
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `filter_even_and_sum` èƒ½å°‡ä»‹æ–¼ `a` èˆ‡ `b` ä¹‹é–“çš„å¶æ•¸æŒ‘å‡ºä¾†åŠ ç¸½å¾Œè¼¸å‡ºï¼Œå…¶ä¸­ $a \leq b$ã€‚

```r
filter_even_and_sum <- function(a, b) {
    # filter_even_and_sum(2, 4)
    # [1] 6
    # filter_even_and_sum(1, 5)
    # [1] 6
    # filter_even_and_sum(2, 10)
    # [1] 30
    # filter_even_and_sum(1, 11)
    # [1] 30
    ### BEGIN SOLUTION
    x <- a:b
    evens <- x[x %% 2 == 0]
    return(sum(evens))
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `filter_negative_and_sum` èƒ½å°‡ `list` ä¸­çš„è² æ•¸æŒ‘å‡ºä¾†åŠ ç¸½å¾Œè¼¸å‡ºã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ¸…å–® `list`ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€‚

```r
filter_negative_and_sum <- function(x) {
    # filter_negative_and_sum(list(-5, 5, -6, 6))
    # [1] -11
    # filter_negative_and_sum(list(-5,-5, -6, -6))
    # [1] -22
    # filter_negative_and_sum(list(5, 5, 6, 6))
    # [1] 0
    # filter_negative_and_sum(list(-5, -5, 6, 6))
    # [1] -10
    # filter_negative_and_sum(list(5, 5, -6, -6))
    # [1] -12
    ### BEGIN SOLUTION
    negatives_sum <- 0
    for (item in x) {
        if (item < 0) {
            negatives_sum <- negatives_sum + item
        }
    }
    return(negatives_sum)
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `get_factors` èƒ½å¤ è¼¸å‡ºä¸€å€‹æ­£æ•´æ•¸çš„æ‰€æœ‰å› æ•¸ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€‚

```r
get_factors <- function(x) {
    # get_factors(1)
    # [1] 1
    # get_factors(2)
    # [1] 1 2
    # get_factors(3)
    # [1] 1 3
    # get_factors(4)
    # [1] 1 2 4
    # get_factors(5)
    # [1] 1 5
    ### BEGIN SOLUTION
    possible_factors <- 1:x
    is_factor <- x %% possible_factors == 0
    factors <- possible_factors[is_factor]
    return(factors)
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `is_prime` èƒ½å¤ åˆ¤æ–·è¼¸å…¥çš„æ­£æ•´æ•¸æ˜¯å¦ç‚ºè³ªæ•¸ï¼ˆæ°å¥½æœ‰å…©å€‹å› æ•¸ 1 èˆ‡è‡ªèº«çš„æ­£æ•´æ•¸ï¼‰ã€‚

- é æœŸè¼¸å‡ºï¼šä¸€å€‹é‚è¼¯å‘é‡ã€‚

```r
is_prime <- function(x) {
    # is_prime(1)
    # [1] FALSE
    # is_prime(2)
    # [1] TRUE
    # is_prime(3)
    # [1] TRUE
    # is_prime(4)
    # [1] FALSE
    # is_prime(5)
    # [1] TRUE
    ### BEGIN SOLUTION
    factors <- get_factors(x)
    n_factors <- length(factors)
    return(n_factors == 2)
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `filter_primes` èƒ½å°‡ä»‹æ–¼ `a` èˆ‡ `b` ä¹‹é–“çš„è³ªæ•¸æŒ‘å‡ºä¾†åŠ ç¸½å¾Œè¼¸å‡ºï¼Œå…¶ä¸­ $a \leq b$ã€‚

- é æœŸè¼¸å…¥ï¼šå…©å€‹æ•¸å€¼å‘é‡ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€‚

```r
filter_primes <- function(a, b) {
    # filter_primes(1, 5)
    # [1] 2 3 5
    # filter_primes(1, 10)
    # [1] 2 3 5 7
    # filter_primes(1, 15)
    # [1] 2 3 5 7 11 13
    ### BEGIN SOLUTION
    out <- c()
    for (i in a:b) {
        if (is_prime(i)) {
            out <- c(out, i)
        }
    }
    return(out)
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `count_vowels` èƒ½è¨ˆç®—è¼¸å…¥è‹±æ–‡å–®å­—ä¸­çš„æ¯éŸ³å€‹æ•¸æœ‰å¹¾å€‹ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ–‡å­—å‘é‡ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€‚

```r
count_vowels <- function(x) {
    # count_vowels("Anakin")
    # [1] 3
    # count_vowels("Luke")
    # [1] 2
    # count_vowels("Skywalker")
    # [1] 2
    # count_vowels("R")
    # [1] 0
    # count_vowels("R Language")
    # [1] 4
    ### BEGIN SOLUTION
    split_characters_list <- strsplit(x, split = "")
    split_characters_vector <- split_characters_list[[1]]
    vowels <- c("a", "e", "i", "o", "u")
    lower_split_characters_vector <- tolower(split_characters_vector)
    out <- 0
    for (char in lower_split_characters_vector) {
        if (char %in% vowels) {
            out <- out + 1
        }
    }
    return(out)
    ### END SOLUTION
}
```

## è‡ªè¡Œå®šç¾©å‡½æ•¸ `reverse_vowels` èƒ½å°‡è¼¸å…¥è‹±æ–‡å–®å­—ä¸­çš„æ¯éŸ³åšå¤§å°å¯«çš„èª¿æ›ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ–‡å­—å‘é‡ã€‚
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡ã€‚

```r
reverse_vowels <- function(x) {
    # reverse_vowels("Anakin")
    # [1] "anAkIn"
    # reverse_vowels("Luke")
    # [1] "LUkE"
    # reverse_vowels("Skywalker")
    # [1] "SkywAlkEr"
    # reverse_vowels("R")
    # [1] "R"
    # reverse_vowels("R Language")
    # [1] "R LAngUAgE"
    ### BEGIN SOLUTION
    split_characters_list <- strsplit(x, split = "")
    split_characters_vector <- split_characters_list[[1]]
    lower_vowels <- c("a", "e", "i", "o", "u")
    upper_vowels <- c("A", "E", "I", "O", "U")
    out <- ""
    for (char in split_characters_vector) {
        if (char %in% lower_vowels) {
            upper_char <- toupper(char)
            out <- paste0(out, upper_char)
        } else if (char %in% upper_vowels) {
            lower_char <- tolower(char)
            out <- paste0(out, lower_char)
        } else {
            out <- paste0(out, char)
        }
    }
    return(out)
    ### END SOLUTION
}
```
